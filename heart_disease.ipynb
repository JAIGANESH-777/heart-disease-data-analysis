```python
import matplotlib.pyplot as plt
import seaborn as sns
# heart disease distribution amnog the patience

sns.countplot(x='target',data=data,hue_order=[1,0],palette=['red','blue'])
label=["Heart Disease","Other Disease"]
plt.xlabel("Heart Disease Status")
plt.ylabel("Number of Patients")
plt.xticks(ticks=[0,1],labels=label)
plt.title("Distribution of Heart Disease")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Age Distribution by Sex and Heart Disease
sns.boxplot(x='sex', y='age', data=data, hue='target')
# x-axis

plt.xlabel('Sex (Male=0, Female=1)') # Informative label with coding scheme
plt.xticks([0, 1], ['Male', 'Female']) # Set custom labels for sex categories
plt.title('Age Distribution by Sex and Heart Disease')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("Heart Disease data.csv")
# The 'cp' feature represents chest pain type with encoded values (0, 1, 2, 3)
chest_pain_labels = {
0: 'sensation of Chest Pain',
1: 'slight Chest Pain',
2: 'mild chest pain',
3: 'severe chest pain'
}
# Analyze key metrics and relationships
# 3. Categorical vs. target variable (heart disease)
sns.countplot(x='cp',hue='target', data=data, order=chest_pain_labels.keys(),
palette=['red','green']) # Consider a color palette for better distinction
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Patients')
plt.title('Chest Pain Type Distribution by Heart Disease')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'age' and 'ca' are column names
sns.histplot(data['age'])
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution')
plt.show()

sns.histplot(data['ca'])
plt.xlabel('Number of Colored Vessels (0-3)')
plt.ylabel('Density')
plt.title('Number of Colored Vessels Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'age' and 'ca' are column names
sns.histplot(x='age', hue='ca', data=data, multiple='dodge', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.title('Distribution of Age by Number of Colored Vessels')
plt.legend(title='Number of Colored Vessels')
plt.show()
```
